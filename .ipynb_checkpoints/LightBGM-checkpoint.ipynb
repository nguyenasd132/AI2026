```markdown
#LightGBM

### 1. Định nghĩa Mô hình / Thuật toán
**LightGBM (Light Gradient Boosting Machine)** là một framework học máy mã nguồn mở dựa trên thuật toán Gradient Boosting Decision Tree (GBDT) do Microsoft phát triển.
Khác với các thuật toán phát triển cây theo mức (level-wise/depth-wise) như XGBoost, LightGBM phát triển cây theo thuật toán chia lá (leaf-wise). Nó sẽ chọn lá có mức giảm sai số lớn nhất để phát triển tiếp. Điều này giúp mô hình hội tụ nhanh hơn và đạt độ chính xác cao hơn.

### 2. Ứng dụng của thuật toán
* **Bài toán Hồi quy (Regression):** Dự đoán giá bất động sản, dự đoán giá cổ phiếu, doanh thu bán lẻ.
* **Bài toán Phân loại (Classification):** Dự đoán khách hàng rời bỏ (Churn prediction), chấm điểm tín dụng (Credit scoring), phát hiện gian lận (Fraud detection).
* **Bài toán Xếp hạng (Ranking):** Các hệ thống gợi ý (Recommender System), công cụ tìm kiếm.

### 3. Điểm mạnh
* **Tốc độ huấn luyện cực nhanh:** Thuật toán sử dụng kỹ thuật Histogram-based để tối ưu hóa việc phân chia nhánh, giúp tăng tốc độ tính toán.
* **Tiết kiệm tài nguyên:** Tiêu tốn rất ít dung lượng bộ nhớ (Memory usage).
* **Xử lý dữ liệu lớn:** Hỗ trợ tốt cho dữ liệu lớn (Big Data) và dữ liệu dạng ma trận thưa (Sparse matrix) như dữ liệu văn bản TF-IDF.
* **Hỗ trợ mặc định:** Tự động xử lý các giá trị khuyết thiếu (missing values) và các biến phân loại (categorical features).

### 4. Điểm yếu
* **Dễ bị quá khớp (Overfitting):** Vì phát triển cây theo chiều lá, độ sâu của cây có thể rất lớn khiến mô hình bị overfit nếu áp dụng trên các tập dữ liệu quá nhỏ (dưới 10.000 mẫu). Cần phải kiểm soát khắt khe thông số `max_depth` và `num_leaves`.
* **Nhiều tham số phức tạp:** Cần kỹ năng để tinh chỉnh (tuning) hàng chục siêu tham số khác nhau để mô hình hoạt động tối ưu.

### 5. Ứng dụng vào giải quyết bài toán dự đoán giá nhà (Bất động sản HCM)
**Lý do lựa chọn LightGBM để giải quyết bài toán này:**
1. **Bản chất dữ liệu:** Dữ liệu bất động sản là dữ liệu dạng bảng (Tabular Data). Thuật toán Tree-based (như LightGBM) được chứng minh là thuật toán thống trị và mang lại kết quả tốt nhất cho loại dữ liệu này.
2. **Khả năng xử lý linh hoạt:** Bài toán yêu cầu trích xuất đặc trưng văn bản (TF-IDF từ mô tả nhà) kết hợp với các chỉ số đo lường (Diện tích, Số phòng). LightGBM xử lý ma trận thưa từ TF-IDF cực kỳ hiệu quả mà không bị tràn RAM.
3. **Chống nhiễu:** Giá nhà đất thường tuân theo phân phối lệch và bị nhiễu bởi các tin đăng ảo. Thuật toán phân nhánh của cây quyết định giúp mô hình bền bỉ, không bị nhạy cảm quá mức trước các giá trị ngoại lai (outliers) so với Linear Regression.

```

````
import lightgbm as lgb
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.feature_extraction.text import TfidfVectorizer
from scipy.sparse import hstack
import re

# ==============================================================================
# BƯỚC 1: ĐỌC VÀ KHÁM PHÁ DỮ LIỆU
# ==============================================================================
print("Đang đọc dữ liệu...")
df = pd.read_csv('mogi_hcm_data.csv')

col_text = 'Title'          # Cột đặc trưng văn bản
col_target = 'Price'        # Cột biến mục tiêu
cols_num = ['Area', 'Bedrooms', 'Bathrooms'] # Các cột đặc trưng số

# Loại bỏ các mẫu thiếu thông tin cốt lõi
df = df.dropna(subset=[col_target, col_text])

# ==============================================================================
# BƯỚC 2: TIỀN XỬ LÝ DỮ LIỆU
# ==============================================================================
print("Đang tiền xử lý dữ liệu...")
def parse_price(price_str):
    if pd.isna(price_str): return np.nan
    price_str = str(price_str).lower().replace(',', '.')
    total = 0.0
    match_ty = re.search(r'([\d\.]+)\s*tỷ', price_str)
    if match_ty: total += float(match_ty.group(1)) * 1000
    match_trieu = re.search(r'([\d\.]+)\s*triệu', price_str)
    if match_trieu: total += float(match_trieu.group(1))
    if total == 0:
        match_num = re.search(r'^([\d\.]+)$', price_str.strip())
        if match_num: return float(match_num.group(1))
        return np.nan
    return total

def parse_number(text):
    if pd.isna(text): return np.nan
    match = re.search(r'([\d\.]+)', str(text).replace(',', '.'))
    if match: return float(match.group(1))
    return np.nan

df[col_target] = df[col_target].apply(parse_price)
df = df.dropna(subset=[col_target])

for col in cols_num:
    if col in df.columns:
        df[col] = df[col].apply(parse_number)
        df[col] = df[col].fillna(df[col].median())

# Loại bỏ ngoại lai (1% thấp nhất và 1% cao nhất)
q_low = df[col_target].quantile(0.01)
q_hi  = df[col_target].quantile(0.99)
df = df[(df[col_target] >= q_low) & (df[col_target] <= q_hi)]

# ==============================================================================
# BƯỚC 3: TRÍCH XUẤT ĐẶC TRƯNG VÀ CHUẨN BỊ DỮ LIỆU
# ==============================================================================
print("Đang trích xuất đặc trưng...")
# Đặc trưng từ khóa
df['is_mat_tien'] = df[col_text].astype(str).str.contains('mặt tiền|mt', case=False, na=False).astype(int)
df['is_hem_xe_hoi'] = df[col_text].astype(str).str.contains('hẻm xe hơi|hxh|ô tô|xe hơi', case=False, na=False).astype(int)
df['is_so_hong'] = df[col_text].astype(str).str.contains('sổ hồng|sổ đỏ|shr|pháp lý', case=False, na=False).astype(int)
df['is_no_hau'] = df[col_text].astype(str).str.contains('nở hậu', case=False, na=False).astype(int)

# TF-IDF
tfidf = TfidfVectorizer(max_features=1500, ngram_range=(1, 3))
X_text = tfidf.fit_transform(df[col_text].astype(str))

cols_num_extended = cols_num + ['is_mat_tien', 'is_hem_xe_hoi', 'is_so_hong', 'is_no_hau']
available_num_cols = [col for col in cols_num_extended if col in df.columns]
X_num = df[available_num_cols].values

# Ghép ma trận đặc trưng
X_combined = hstack([X_num, X_text]).tocsr()
y = np.log1p(df[col_target].values)

X_train, X_test, y_train, y_test = train_test_split(X_combined, y, test_size=0.2, random_state=42)

# ==============================================================================
# BƯỚC 4: HUẤN LUYỆN MÔ HÌNH LIGHTGBM
# ==============================================================================
print("Bắt đầu huấn luyện mô hình LightGBM...")
train_data = lgb.Dataset(X_train, label=y_train)
test_data = lgb.Dataset(X_test, label=y_test, reference=train_data)

params = {
    'objective': 'regression',
    'metric': 'rmse',
    'boosting_type': 'gbdt',
    'learning_rate': 0.03,
    'num_leaves': 63,
    'max_depth': -1,
    'feature_fraction': 0.8,
    'bagging_fraction': 0.8,
    'bagging_freq': 5,
    'lambda_l1': 1.5,
    'lambda_l2': 1.5,
    'random_state': 42,
    'verbose': -1
}

model = lgb.train(
    params,
    train_data,
    num_boost_round=5000,
    valid_sets=[train_data, test_data],
    callbacks=[
        lgb.early_stopping(stopping_rounds=50),
        lgb.log_evaluation(period=100)
    ]
)

# ==============================================================================
# BƯỚC 5: ĐÁNH GIÁ MÔ HÌNH
# ==============================================================================
y_pred_log = model.predict(X_test, num_iteration=model.best_iteration)
y_pred = np.expm1(y_pred_log)
y_test_real = np.expm1(y_test)

rmse = np.sqrt(mean_squared_error(y_test_real, y_pred))
mae = mean_absolute_error(y_test_real, y_pred)
r2 = r2_score(y_test_real, y_pred)

print(f"\n--- Kết quả đánh giá LightGBM ---")
print(f"RMSE: {rmse:.4f} (Lệch khoảng {rmse:.0f} triệu)")
print(f"MAE:  {mae:.4f} (Lệch khoảng {mae:.0f} triệu)")
print(f"R2:   {r2:.4f} (Độ chính xác tương đối: {r2*100:.2f}%)")
````
