SVR

1.Định nghĩa SVR

Support Vector Machine (SVM) là một thuật toán Học máy có giám sát (Supervised Learning). Dù có thể dùng cho cả dự đoán số liệu, nhưng sức mạnh lớn nhất của SVM là giải quyết bài toán Phân loại (Classification).
Nhiệm vụ duy nhất của SVM là đi tìm một đường ranh giới (hoặc mặt phẳng) tốt nhất để chia cắt các nhóm dữ liệu khác nhau một cách rõ ràng nhất."

2.Nguyên tắc hoạt động SVR
Nguyên tắc của SVM không chỉ là tìm ra một ranh giới để chia cắt dữ liệu, mà ranh giới đó phải an toàn nhất. Nghĩa là khoảng cách từ ranh giới đó đến những điểm dữ liệu gần nhất của các nhóm phải là lớn nhất có thể."

3. Cách thức hoạt động
Hyperplane (Siêu mặt phẳng - Đường ranh giới):
*   Hyperplane (Siêu mặt phẳng - Đường ranh giới): Nếu dữ liệu nằm trên tờ giấy 2D, nó là một đường thẳng. Nếu dữ liệu trong không gian 3D, nó là một mặt phẳng. Đây chính là "nhát cắt" quyết định để chia tách các nhóm. Mọi dữ liệu mới rơi vào bên trái nhát cắt này sẽ thuộc Nhóm A, rơi vào bên phải sẽ thuộc Nhóm B.
*   Support Vectors (Các vector hỗ trợ): Đây là những điểm dữ liệu quan trọng nhất. Chúng là những điểm nằm sát với ranh giới nhất. Mô hình SVM thực chất chỉ được xây dựng dựa trên vị trí của các điểm "mũi nhọn" này, còn các điểm nằm sâu bên trong an toàn thì mô hình không quan tâm.
*   Margin (Khoảng cách lề): Là vùng đệm giới hạn bởi các Support Vectors. SVM sẽ liên tục xoay và dịch chuyển Hyperplane sao cho cái Margin này đạt độ rộng tối đa. Margin càng rộng, mô hình càng có khả năng phân loại chính xác dữ liệu mới trong thực tế.

4. Điểm mạnh và điểm yếu

a. điểm mạnh

*   Hiệu suất cao trong không gian đặc trưng đa chiều (High-dimensional spaces): Đối với các bài toán yêu cầu xử lý ngôn ngữ tự nhiên (NLP) thông qua việc trích xuất đặc trưng từ văn bản mô tả (sử dụng TF-IDF hoặc Word Embedding), tập dữ liệu đầu vào thường có số lượng chiều (features) rất lớn. Họ thuật toán SVM/SVR thể hiện sự vượt trội trong việc tính toán và xây dựng siêu mặt phẳng (Hyperplane) tối ưu trong các không gian này mà không bị suy giảm hiệu suất nghiêm trọng (tránh được hiện tượng Lời nguyền đa chiều - Curse of Dimensionality).
*   Khả năng kháng nhiễu (Robustness to Outliers): Dữ liệu thu thập (crawl) từ các trang rao vặt bất động sản thực tế thường chứa nhiều nhiễu và các giá trị ngoại lai (outliers) do sai sót trong quá trình nhập liệu hoặc các tin đăng ảo. Thông qua cơ chế tối ưu hóa chỉ phụ thuộc vào các điểm dữ liệu tới hạn (Support Vectors) kết hợp với khoảng dung sai cho phép, SVR có khả năng bỏ qua các dao động bất thường, giúp mô hình nắm bắt được xu hướng giá cốt lõi của thị trường ổn định hơn.


*   Khả năng mô hình hóa dữ liệu phi tuyến tính: Giá trị bất động sản hiếm khi biến thiên tuyến tính theo các đặc trưng độc lập. Thông qua thủ thuật hạt nhân (Kernel Trick), tiêu biểu là hàm Radial Basis Function (RBF), SVR có thể ánh xạ dữ liệu vào không gian nhiều chiều hơn, từ đó phát hiện và mô hình hóa các mối quan hệ phức tạp, phi tuyến tính giữa các biến số (diện tích, cấu trúc, vị trí) và giá nhà.

b. điểm yếu



*   Hạn chế về khả năng mở rộng quy mô (Scalability): Độ phức tạp tính toán của thuật toán tỷ lệ thuận theo cấp số phương so với số lượng mẫu huấn luyện. Mặc dù SVR hoạt động cực kỳ hiệu quả với tập dữ liệu quy mô nhỏ và trung bình (khoảng 5.000 mẫu tin như yêu cầu của đồ án), thuật toán sẽ đòi hỏi tài nguyên tính toán lớn và thời gian huấn luyện kéo dài đáng kể nếu kích thước tập dữ liệu tăng lên hàng chục hoặc hàng trăm ngàn mẫu.
*   Đòi hỏi tinh chỉnh tham số (Hyperparameter Tuning): Hiệu năng dự báo của SVR phụ thuộc rất lớn vào việc lựa chọn đúng bộ siêu tham số, bao gồm loại hàm Kernel, tham số điều kiện,.... Quá trình tinh chỉnh này đòi hỏi áp dụng các kỹ thuật tìm kiếm như Grid Search hoặc Random Search, làm tăng thời gian xây dựng mô hình.









```python
# Xây dựng lớp mô hình SVM Phân loại đơn giản
class SimpleLinearSVM:
    # 1. Hàm khởi tạo "bộ khung" và "luật chơi"
    def __init__(self, learning_rate=0.001, epochs=1000, lambda_param=0.01):
        # Bộ não (w1: trọng số diện tích, w2: trọng số giá, b: độ lệch)
        self.w1 = 0.0
        self.w2 = 0.0
        self.b = 0.0

        # Luật chơi
        self.learning_rate = learning_rate
        self.epochs = epochs
        self.lambda_param = lambda_param

    # 2. Hàm huấn luyện (Máy tính đọc dữ liệu và tự học)
    def fit(self, X_dien_tich, X_gia, y):
        n = len(y) # Đếm số lượng căn nhà

        # Vòng lặp ngoài: Lặp lại quá trình học (Epochs)
        for epoch in range(self.epochs):
            # Vòng lặp trong: Cầm từng căn nhà lên xem xét
            for i in range(n):

                # --- [BẮT ĐẦU VÀO LÕI CỦA SVM] ---

                # Bước A: Tính điểm xem đoán đúng hay sai, cách ranh giới bao xa
                condition = y[i] * (self.w1 * X_dien_tich[i] + self.w2 * X_gia[i] + self.b)

                # Bước B: Cơ chế thưởng phạt (Cốt lõi làm nên đường ranh giới)
                if condition >= 1:
                    # TRẠNG THÁI AN TOÀN: Đoán đúng và nằm ngoài vùng đệm.
                    # Chỉ tinh chỉnh nhẹ để làm phẳng ranh giới (KHÔNG dùng dữ liệu căn nhà)
                    self.w1 = self.w1 - self.learning_rate * (2 * self.lambda_param * self.w1)
                    self.w2 = self.w2 - self.learning_rate * (2 * self.lambda_param * self.w2)
                else:
                    # TRẠNG THÁI BỊ PHẠT: Đoán sai hoặc dẫm lên vùng đệm.
                    # Bắt buộc kéo ranh giới về phía căn nhà này (CÓ DÙNG thông tin X và y của nhà)
                    self.w1 = self.w1 - self.learning_rate * (2 * self.lambda_param * self.w1 - X_dien_tich[i] * y[i])
                    self.w2 = self.w2 - self.learning_rate * (2 * self.lambda_param * self.w2 - X_gia[i] * y[i])
                    self.b  = self.b  - self.learning_rate * (-y[i])

                # --- [KẾT THÚC LÕI SVM] ---

    # 3. Hàm dự đoán
    def predict(self, dien_tich, gia):
        result = self.w1 * dien_tich + self.w2 * gia + self.b
        if result > 0:
            return 1  # TP.HCM
        else:
            return -1 # Hà Nội

# --- CHẠY THỬ CHƯƠNG TRÌNH ---
# Dữ liệu giả lập
X_dien_tich = [0.5, 0.6, 0.45, 0.9, 1.2, 1.5] # m2 (đã chia 100)
X_gia       = [6.5, 8.0, 5.0,  3.5, 4.0, 4.5] # Tỷ VNĐ
y_khu_vuc   = [1,   1,   1,   -1,  -1,  -1]   # 1: HCM, -1: HN

# Khởi tạo và huấn luyện
model = SimpleLinearSVM(learning_rate=0.01, epochs=1000, lambda_param=0.01)
print("Đang huấn luyện mô hình SVM...")
model.fit(X_dien_tich, X_gia, y_khu_vuc)

# Dự đoán thử
dt_moi, gia_moi = 0.55, 7.0
ket_qua = model.predict(dt_moi, gia_moi)

print(f"Nhà 55m2, giá 7 Tỷ VND nằm ở: {'TP.HCM' if ket_qua == 1 else 'Hà Nội'}")
```

**Cơ sở chuyển đổi từ SVM sang SVR cho bài toán đồ án:**
Mặc dù Support Vector Machine (SVM) là thuật toán cốt lõi, nhưng việc áp dụng trực tiếp SVM vào đồ án dự báo giá nhà là không phù hợp do sự khác biệt về mục tiêu đầu ra. Cụ thể:


*   Bản chất của SVM: SVM nguyên thủy được thiết kế để giải quyết bài toán Phân loại (Classification). Mục tiêu của thuật toán là tìm kiếm một siêu mặt phẳng tối ưu nhằm phân định dữ liệu thành các lớp (Classes) rời rạc. Nếu áp dụng SVM, hệ thống chỉ có thể dự báo các nhãn phân loại (ví dụ: nhà phân khúc Giá rẻ, Trung cấp, hoặc Cao cấp).
*   Lý do ứng dụng SVR: Yêu cầu kỹ thuật của hệ thống là dự báo giá nhà dưới dạng một biến số liên tục (định lượng cụ thể). Do đó, bài toán được định nghĩa là bài toán Hồi quy (Regression). Support Vector Regression (SVR) là phiên bản mở rộng của SVM, được tinh chỉnh lại hàm mục tiêu toán học. Thay vì tìm ranh giới chia cắt dữ liệu, SVR cố gắng xây dựng một đường dự báo cốt lõi sao cho phần lớn các điểm dữ liệu thực tế nằm gọn bên trong một khoảng sai số cho phép (Hàm mất mát ϵ-insensitive). Sự chuyển đổi cơ chế này cho phép kế thừa toàn bộ ưu điểm toán học của SVM để giải quyết bài toán dự báo giá trị liên tục trong lĩnh vực bất động sản.
